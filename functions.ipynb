```python
def converter(ForC,d ):
    if ForC == "F":
        return((d - 32) * 1.8)
    elif ForC == "C":
        return(1.8*d + 32.0)
ForC = input("""What do you want to convert?
Choose F if you want to convert Fahrenheit to Celsius
Choose C if you want to convert Celcius to Fahrenheit
""" )
d = int(input("What is the temperature: "))
m = converter(ForC, d)
if ForC =="F":
    print(d,"degrees Fahrentein equals to ",m, "degrees Celsius")
elif ForC == "C":
          print(d,"degrees Celsius equals to ",m,"degrees Fahrentein")
```

```python
def quadequat(a,b,c):
    x1=(-b+(b**2-4*a*c)**(1/2))/2*a
    x2=(-b+(b**2-4*a*c)**(1/2))/2*a
    return(x1,x2)
quadequat(1,2,1)
```

```python
def geom_prog(b1,q):
    if abs(q)<1:
        S=b1/(1-q)
        print("The sum of this progression is ",S)
    else:
        print("|q| should be less than 1")
geom_prog(1,(-1/2)) 
```
