```python
current_users=['Jason','Alex','Wayne','Bruce','Kim']
new_users=['Jim','Johny','ALEX','Bruce','Stephen']

new_list =[]

for current_user in current_users:
    new_list.append(current_user.lower())

for new_user in new_users:
    if new_user.lower() in new_list:
        print ("This username, " + new_user + ", is taken")
    else:
        print ("Feel free to use " + new_user)
```
